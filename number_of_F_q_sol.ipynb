```python
import itertools

def q1(a, b, c, d, e, f):
    return (a*c*d*f - a*b*d - a*c*e - b*c*f - d*e*f 
            + a**2 + b**2 + c**2 + d**2 + e**2 + f**2)

def q2(a, b, c, d, e, f):
    return a*f - b*e + c*d

def count_solutions(p):
    count = 0
    # naive approach: O(p^6) - slow for p > 11
    # 最適化の余地はありますが、まずはこれで小さなpを見るのが確実です
    for a, b, c, d, e, f in itertools.product(range(p), repeat=6):
        # q2 = 4 (mod p) check
        if (q2(a, b, c, d, e, f) + 4) % p != 0:
            continue
        
        # q1 = 8 (mod p) check
        if (q1(a, b, c, d, e, f) - 8) % p == 0:
            count += 1
            
    return count

primes = [2, 3, 5, 7, 11, 13, 19, 23, 29] # p=13以降は計算時間と相談
print(f"p | N_p (q2=-4) | Total approx (x2)")
print("-" * 30)
for p in primes:
    n = count_solutions(p)
    total = n * 2 if p != 2 else n # p=2の場合は+4と-4は同じ
    print(f"{p} | {n:^10} | {total:^12}")
```
